```python
import numpy as np
from ase import Atoms
from ase.visualize import view
```

# Ag adatom on Ni slab

```python
a = 3.55
atoms = Atoms("Ni4",
              cell=[np.sqrt(2) * a,
                    np.sqrt(2) * a, 1.0, 90, 90, 120],
              pbc=(1, 1, 0),
              scaled_positions=[(0, 0, 0), (0.5, 0, 0), (0, 0.5, 0),
                                (0.5, 0.5, 0)])
atoms.center(vacuum=5.0, axis=2)
h = 1.9
relative = (1 / 6, 1 / 6, 0.5)
absolute = np.dot(relative, atoms.cell) + (0, 0, h)
atoms.append('Ag')
atoms.positions[-1] = absolute
view(atoms.repeat(5), viewer='x3d')
```
